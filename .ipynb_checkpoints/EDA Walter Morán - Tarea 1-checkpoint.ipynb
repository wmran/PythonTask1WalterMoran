### **2. Limpieza y transformación de los datos**

Una vez que tengas el conjunto de datos, el siguiente paso es limpiarlo y transformarlo para que esté listo para el análisis. Recuerda verificar las variables, gestionar los valores faltantes, corregir errores y similares

```python
# Importacion de la libreria de pandas
import pandas as pd

#Importación y lectura del dataset Impact_of_Remote_Work_on_Mental_Health.csv
df = pd.read_csv('./data/Impact_of_Remote_Work_on_Mental_Health.csv')
```

```python
# Visualización de las primeras 5 lineas del dataframe
df.head()
```

```python
# Extracción de columnas a utilizar en nuestro EDA
df = df.loc[:,["Employee_ID"
               ,"Age"
               ,"Gender"
               ,"Job_Role"
               ,"Industry"
               ,"Years_of_Experience"
               ,"Work_Location"
               ,"Hours_Worked_Per_Week"
               ,"Number_of_Virtual_Meetings"
               ,"Stress_Level"
               ,"Mental_Health_Condition"
               ,"Access_to_Mental_Health_Resources"
               ,"Productivity_Change"
               ,"Satisfaction_with_Remote_Work"
               ,"Physical_Activity"]]
```

```python
# Cantidad de Filas y Columnas actuales
df.shape
```

```python
# Renombre de columnas para mayor comprension
df.rename(columns={"Employee_ID":"IDEmpleado",
                   "Age":"Edad",
                   "Gender":"Genero",
                   "Job_Role":"Rol",
                   "Industry":"Industria",
                   "Years_of_Experience":"AniosExperiencia",
                   "Work_Location":"Locacion",
                   "Hours_Worked_Per_Week":"HorasTrabajadasSemana",
                   "Number_of_Virtual_Meetings":"CantidadReunionesVirtuales",
                   "Stress_Level":"NivelEstres",
                   "Mental_Health_Condition":"CondicionMental",
                   "Access_to_Mental_Health_Resources":"AccesoRecursosSaludMental",
                   "Productivity_Change":"CambioProductividad",
                   "Satisfaction_with_Remote_Work":"SatisfaccionTrabajoRemoto",
                   "Physical_Activity":"ActividadFisica"},
          inplace=True)
```

```python
#Creación de Indice de Dataframe
df = df.set_index("IDEmpleado")
```

```python
# Columnas del Dataframe
df.columns
```

```python
# Información de tipos de datos por columnas
df.dtypes
```

```python
# Visualización de las primeras 5 lineas del DataFrame actual
df.head()
```

```python
# Conteo de registros nulos dentro del Dataframe
df.isnull().sum()
```

```python
# Limpieza de Nulos en Dataset
df.dropna(inplace=True)
```

```python
# Conteo de registros nulos dentro del Dataframe
df.isnull().sum()
```

```python
# Cantidad de Filas y Columnas actuales
df.shape
```

```python
# Asignación del Dataframe limpio y transformado en nuevo archivo para mantener 
# una buena organización del dataset original y el dataset utilizado en nuestra investigacion
df_limpio = df
```

```python
# Creando nuevo archivo CSV con la data limpia y con las columnas las cuales tomaremos en cuenta para nuestro calculo
df_limpio.to_csv('./data/Impact_of_Remote_Work_on_Mental_Health_train_test.csv', index=False)
```

### **4. Visualización de resultados**

###### Posteriormente ejecuta la visualización de los datos para comunicar hallazgos y comprender mejor las características del conjunto de datos. Puedes apoyarte de un histograma, un gráfico de barras, de pastel o cualquier otro gráfico que resulte adecuado para los datos seleccionados.

```python
%matplotlib inline
```

```python
# Histograma de columna edad
sns.histplot(df['Edad'])
plt.title('Histograma de Edad')
plt.show()
```

```python
# Conversion de variables
df_CamposNumericos = df.select_dtypes(include=['float64', 'int64'])

# Matriz de correlacion
correlacion = df_CamposNumericos.corr()
sns.heatmap(correlacion, cmap='coolwarm')
plt.show()
```

```python
# Grafica de Niveles de estres por Locacion
df_EstresxLocacion = df.groupby(['Locacion', 'NivelEstres']).size()

df_EstresxLocacion.plot(kind='bar')
plt.title('Conteo de Niveles de Estrés por Locación')
plt.ylabel('Cantidad de Empleados')
plt.xlabel('Locación')
plt.legend(title='Nivel de Estrés')
plt.show()
```

```python
# Conteo de Personas que realizan trabajo remoto según su Genero
df_Remoto = df[df['Locacion'] == 'Remote']
df_Remoto_count = df_Remoto['Genero'].value_counts()

df_Remoto_count.plot(kind='pie', autopct='%1.1f%%')
plt.title('Cantidad de Personas que Realizan Trabajo Remoto según Género')
plt.ylabel('Cantidad de Personas')
plt.show()
```

```python
# Conteo de Personas que realizan trabajo remoto según su Genero
df_EnSitio = df[df['Locacion'] == 'Onsite']
df_EnSitio_count = df_EnSitio['Genero'].value_counts()

df_EnSitio_count.plot(kind='pie', autopct='%1.1f%%')
plt.title('Cantidad de Personas que Realizan Trabajo En sitio según Género')
plt.ylabel('Cantidad de Personas')
plt.show()
```

```python
# Conteo de Personas que realizan trabajo remoto según su Genero
df_Hybrid = df[df['Locacion'] == 'Hybrid']
df_Hybrid_count = df_Hybrid['Genero'].value_counts()

df_Hybrid_count.plot(kind='pie', autopct='%1.1f%%')
plt.title('Cantidad de Personas que Realizan Trabajo Hibrido según Género')
plt.ylabel('Cantidad de Personas')
plt.show()
```

```python
# Contar la cantidad de personas por nivel de estrés
stress_count = df['NivelEstres'].value_counts()

# Graficar
stress_count.plot(kind='bar', figsize=(10, 6))
plt.title('Cantidad de Personas por Nivel de Estrés')
plt.ylabel('Cantidad de Personas')
plt.xlabel('Nivel de Estrés')
plt.xticks(rotation=0)
plt.show()
```
